```python
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
wine_df = pd.read_csv('C:/Users/Vigoursoft/wine.data.csv', names = ["Cultivator", "Alcohol", "Malic_acid", "Ash", "Alcalinity_of_Ash", "Magnesium", "Total_phenols", "Falvanoids", "Nonflavanoid_phenols", "Proanthocyanins", "Color_intensity", "Hue","00280", "Proline"])
```

```python
from sklearn.model_selection import train_test_split

X = wine_df.drop("Cultivator", axis=1)
y = wine_df.pop("Cultivator")
```

```python
X.head()
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
model = GaussianNB()
model.fit(train_set, train_labels)
```

```python
model.score(train_set, train_labels)
```

```python
test_pred = model.predict(test_set)

print(metrics.classification_report(test_labels, test_pred))
print(metrics.confusion_matrix(test_labels, test_pred))
```
